NODE 2 -> Router

```python
def router_node(state:QuantumState) -> dict:
    question=state["question"]
    messages=state.get("messages",[])
    recent   = "; ".join(
        f"{m['role']}: {m['content'][:60]}"
        for m in messages[-3:-1]
    ) or "none"

    prompt = f"""You are a router for a quantum physics study assistant.
    Available options:
- retrieve: student is asking about a quantum physics concept, theory,
  history, scientist, or phenomenon that can be answered from the
  knowledge base. Examples: "What is superposition?", "Who was Bohr?",
  "Explain the uncertainty principle", "How does tunnelling work?"

- tool: student is asking a numerical problem that requires calculation.
  Use this when the question contains numbers, units, or asks to
  calculate, find, or compute something. Examples: "What is the energy
  of a photon at 6×10¹⁴ Hz?", "Calculate the de Broglie wavelength of
  an electron moving at 2×10⁶ m/s", "Find the hydrogen energy at n=3"

- memory_only: student is asking a follow-up question about the
  conversation itself, asking what was just said, asking what topics
  were covered, or making casual small talk. Examples: "What did you
  just say?", "What topics have we covered?", "My name is Arjun",
  "Can you repeat that?", "Thanks!"

Recent conversation: {recent}
Current question: {question}
Reply with ONLY one word: retrieve / memory_only / tool"""

    response = llm.invoke(prompt)
    decision = response.content.strip().lower()

    # Clean up LLM output — handle unexpected responses
    if "memory" in decision:    decision = "memory_only"
    elif "tool" in decision:    decision = "tool"
    else:                       decision = "retrieve"

    print(f"  [router] route='{decision}' for: {question[:50]}")
    return {"route": decision}

test_1 = {
    "question": "What did you just say?",
    "messages": [{"role": "user", "content": "hi"}]
}
result_1 = router_node(test_1)
print(f"Test 1: route='{result_1['route']}' (expected: memory_only)")
test_2 = {
    "question": "What is the uncertainty principle?",
    "messages": []
}
result_2 = router_node(test_2)
print(f"Test 2: route='{result_2['route']}' (expected: retrieve)")
test_3 = {
    "question": "What is the energy of a photon at 6×10¹⁴ Hz?",
    "messages": []
}
result_3 = router_node(test_3)
print(f"Test 3: route='{result_3['route']}' (expected: tool)")
test_4 = {
    "question": "Who was Niels Bohr and what did he discover?",
    "messages": []
}
result_4 = router_node(test_4)
print(f"Test 4: route='{result_4['route']}' (expected: retrieve)")
test_5 = {
    "question": "Calculate de Broglie wavelength of electron at 2×10⁶ m/s",
    "messages": []
}
result_5 = router_node(test_5)
print(f"Test 5: route='{result_5['route']}' (expected: tool)")
test_6 = {
    "question": "What topics have we covered so far?",
    "messages": [
        {"role": "user",      "content": "What is superposition?"},
        {"role": "assistant", "content": "Superposition is..."}
    ]
}
result_6 = router_node(test_6)
print(f"Test 6: route='{result_6['route']}' (expected: memory_only)")
print("\n✅ router_node tests complete")
```

NODE 3 -> Retrieval node

```python
def retrieval_node(state: QuantumState) -> dict:
    q_emb   = embedder.encode([state["question"]]).tolist()
    results = collection.query(query_embeddings=q_emb, n_results=3)
    chunks  = results["documents"][0]
    topics  = [m["topic"] for m in results["metadatas"][0]]
    context = "\n\n---\n\n".join(
        f"[{topics[i]}]\n{chunks[i]}"
        for i in range(len(chunks))
    )
    print(f"  [retrieval] sources={topics}")
    return {"retrieved": context, "sources": topics}

def skip_retrieval_node(state:QuantumState) -> dict:
    print(f" [skip] no retrieval needed")
    return {"retrieved":"","sources":[]}

test_1 = {"question": "What is the uncertainty principle?"}
result_1 = retrieval_node(test_1)
print(f"\nTest 1 — Concept question:")
print(f"  sources: {result_1['sources']}")
print(f"  preview: {result_1['retrieved'][:150]}...")
test_2 = {"question": "Who was Niels Bohr?"}
result_2 = retrieval_node(test_2)
print(f"\nTest 2 — Scientist question:")
print(f"  sources: {result_2['sources']}")
print(f"  preview: {result_2['retrieved'][:150]}...")
test_3 = {"question": "What is the de Broglie wavelength formula?"}
result_3 = retrieval_node(test_3)
print(f"\nTest 3 — Formula question:")
print(f"  sources: {result_3['sources']}")
print(f"  preview: {result_3['retrieved'][:150]}...")
result_4 = skip_retrieval_node({})
print(f"\nTest 4 — Skip node:")
print(f"  retrieved: '{result_4['retrieved']}'")
print(f"  sources: {result_4['sources']}")

print("\n✅ retrieval_node and skip_retrieval_node tests complete")
```

NODE 4 -> TOOL NODE

```python
import re
import math
PLANCK_H = 6.626e-34
PLANCK_HBAR = 1.055e-34
SPEED_LIGHT = 3.0e8    
ELECTRON_MASS = 9.109e-31
ELECTRON_VOLT = 1.602e-19
def quantum_calculator(question:str) -> str:
    """
    Detects what calculation is needed and solve it 
    step by step. Return a formatted string always
    Never raises a exception
    """
    q=question.lower()
    try:
        if ("photon energy" in q or "energy of photon" in q or ("energy" in q and "frequency" in q and "photon" in q)):
            freq = extract_number(question)
            if freq is None:
                return "please provide the frequency value in Hz."
            energy_j=PLANCK_H*freq
            energy_ev=energy_j/ELECTRON_VOLT
            return f"""📐 Photon Energy Calculation (E = hf)

Formula:  E = h × f
Where:    h = 6.626 × 10⁻³⁴ J·s (Planck's constant)
          f = {freq:.3e} Hz

Step 1:   E = 6.626 × 10⁻³⁴ × {freq:.3e}
Step 2:   E = {energy_j:.3e} J
Step 3:   E = {energy_ev:.3e} eV

Result:   E = {energy_j:.3e} J = {energy_ev:.3e} eV"""
        elif("energy" in q and "wavelength" in q and "photon" in q):
            wavelength = extract_number(question)
            if wavelength is None:
                return "Please provide the wavelength value in metres."
            energy_j  = (PLANCK_H * SPEED_LIGHT) / wavelength
            energy_ev = energy_j / ELECTRON_VOLT
            return f"""📐 Photon Energy from Wavelength (E = hc/λ)
Formula: E = hc / λ
Where: h = 6.626 × 10⁻³⁴ J·s
          c = 3.0 × 10⁸ m/s
          λ = {wavelength:.3e} m
Step 1: E = (6.626 × 10⁻³⁴ × 3.0 × 10⁸) / {wavelength:.3e}
Step 2: E = {PLANCK_H * SPEED_LIGHT:.3e} / {wavelength:.3e}
Step 3: E = {energy_j:.3e} J
Result: E = {energy_j:.3e} J = {energy_ev:.3e} eV"""
        
        elif ("de broglie" in q or "matter wave" in q
              or ("wavelength" in q and
                  any(p in q for p in
                      ["electron","proton","particle","neutron"]))):
            velocity = extract_number(question)
            if velocity is None:
                return "Please provide the velocity value in m/s."
            mass = ELECTRON_MASS
            if "proton" in q:
                mass = 1.673e-27
            elif "neutron" in q:
                mass = 1.675e-27
            wavelength = PLANCK_H / (mass * velocity)
            particle   = ("proton" if "proton" in q
                         else "neutron" if "neutron" in q
                         else "electron")
            return f""" de Broglie Wavelength (λ = h/mv)
Formula:  λ = h / (m × v)
Where:    h = 6.626 × 10⁻³⁴ J·s
          m = {mass:.3e} kg ({particle} mass)
          v = {velocity:.3e} m/s
Step 1:   m × v = {mass:.3e} × {velocity:.3e}
Step 2:   m × v = {mass * velocity:.3e} kg·m/s
Step 3:   λ = 6.626 × 10⁻³⁴ / {mass * velocity:.3e}
Result:   λ = {wavelength:.3e} m"""
        elif (("energy level" in q or "hydrogen" in q)
              and any(f"n={i}" in q.replace(" ","")
                     for i in range(1, 11))):
            n = None
            for i in range(1, 11):
                if f"n={i}" in q.replace(" ", ""):
                    n = i
                    break

            if n is None:
                return "Please specify the energy level n (e.g. n=3)."

            energy_ev = -13.6 / (n ** 2)
            energy_j  = energy_ev * ELECTRON_VOLT
            return f"""📐 Hydrogen Energy Level (Eₙ = -13.6/n² eV)

Formula:  Eₙ = -13.6 / n² eV
Where:    n = {n} (principal quantum number)

Step 1:   E{n} = -13.6 / {n}²
Step 2:   E{n} = -13.6 / {n**2}
Step 3:   E{n} = {energy_ev:.4f} eV

Result:   E{n} = {energy_ev:.4f} eV = {energy_j:.3e} J

Note: Negative sign means electron is bound to nucleus.
      Higher n = higher energy = less tightly bound."""
        elif ("transition" in q or "jump" in q
              or ("from n" in q and "to n" in q)):

            numbers = re.findall(r'n\s*=\s*(\d+)', q)
            if len(numbers) < 2:
                numbers = re.findall(r'\d+', question)

            if len(numbers) < 2:
                return ("Please specify both energy levels "
                        "e.g. 'from n=3 to n=2'")

            n_upper = int(numbers[0])
            n_lower = int(numbers[1])

            if n_upper < n_lower:
                n_upper, n_lower = n_lower, n_upper

            e_upper = -13.6 / (n_upper ** 2)
            e_lower = -13.6 / (n_lower ** 2)
            delta_e = abs(e_upper - e_lower)
            freq    = (delta_e * ELECTRON_VOLT) / PLANCK_H
            wavel   = SPEED_LIGHT / freq

            return f"""📐 Energy Level Transition (ΔE = Eupper - Elower)

Formula:  ΔE = |E_upper - E_lower|

Step 1:   E{n_upper} = -13.6 / {n_upper}² = {e_upper:.4f} eV
Step 2:   E{n_lower} = -13.6 / {n_lower}² = {e_lower:.4f} eV
Step 3:   ΔE = |{e_upper:.4f} - {e_lower:.4f}|
Step 4:   ΔE = {delta_e:.4f} eV

Photon frequency: f = {freq:.3e} Hz
Photon wavelength: λ = {wavel:.3e} m = {wavel*1e9:.1f} nm"""
        elif ("photoelectric" in q or "work function" in q
              or ("kinetic energy" in q and
                  any(w in q for w in ["photon","light","metal"]))):

            numbers = extract_all_numbers(question)
            if len(numbers) < 2:
                return ("Please provide both frequency (Hz) "
                        "and work function (eV).")

            freq    = numbers[0]
            phi_ev  = numbers[1]
            phi_j   = phi_ev * ELECTRON_VOLT
            ke_j    = PLANCK_H * freq - phi_j
            ke_ev   = ke_j / ELECTRON_VOLT

            if ke_j < 0:
                return f"""📐 Photoelectric Effect (KE = hf - Φ)
            Formula:  KE = hf - Φ
          hf = {PLANCK_H * freq:.3e} J
          Φ  = {phi_j:.3e} J ({phi_ev} eV)

Result:   KE = {ke_j:.3e} J (NEGATIVE)

⚠️ No electrons ejected — photon energy is less
   than the work function. Frequency is below
   the threshold frequency for this metal."""

            return f"""📐 Photoelectric Effect (KE = hf - Φ)

Formula:  KE = hf - Φ
Where:    h = 6.626 × 10⁻³⁴ J·s
          f = {freq:.3e} Hz
          Φ = {phi_ev} eV = {phi_j:.3e} J
          Step 1:   hf = 6.626 × 10⁻³⁴ × {freq:.3e} = {PLANCK_H*freq:.3e} J
Step 2:   KE = {PLANCK_H*freq:.3e} - {phi_j:.3e}
Step 3:   KE = {ke_j:.3e} J

Result:   KE = {ke_j:.3e} J = {ke_ev:.4f} eV
          Electrons ARE ejected from the metal."""
        elif ("uncertainty" in q
              or "delta x" in q or "Δx" in question
              or "delta p" in q or "Δp" in question):

            number = extract_number(question)
            if number is None:
                return "Please provide a value for position or momentum uncertainty."

            if ("position" in q or "delta x" in q
                    or "Δx" in question):
                delta_x  = number
                delta_p  = PLANCK_HBAR / (2 * delta_x)
                return f"""📐 Heisenberg Uncertainty Principle (Δx·Δp ≥ ℏ/2)
            Formula:  Δp ≥ ℏ / (2 × Δx)
Where:    ℏ  = 1.055 × 10⁻³⁴ J·s
          Δx = {delta_x:.3e} m

Step 1:   Δp ≥ 1.055 × 10⁻³⁴ / (2 × {delta_x:.3e})
Step 2:   Δp ≥ 1.055 × 10⁻³⁴ / {2*delta_x:.3e}

Result:   Δp ≥ {delta_p:.3e} kg·m/s (minimum momentum uncertainty)"""
            else:
                delta_p  = number
                delta_x  = PLANCK_HBAR / (2 * delta_p)
                return f"""📐 Heisenberg Uncertainty Principle (Δx·Δp ≥ ℏ/2)
Formula:  Δx ≥ ℏ / (2 × Δp)
Where:    ℏ  = 1.055 × 10⁻³⁴ J·s
          Δp = {delta_p:.3e} kg·m/s
Step 1:   Δx ≥ 1.055 × 10⁻³⁴ / (2 × {delta_p:.3e})
Result:   Δx ≥ {delta_x:.3e} m (minimum position uncertainty)"""
        elif ("frequency" in q and "wavelength" in q
              and "photon" not in q):
            wavelength = extract_number(question)
            if wavelength is None:
                return "Please provide the wavelength in metres."
            freq = SPEED_LIGHT / wavelength
            return f"""📐 Frequency from Wavelength (f = c/λ)
Formula:  f = c / λ
Where:    c = 3.0 × 10⁸ m/s
          λ = {wavelength:.3e} m
Step 1:   f = 3.0 × 10⁸ / {wavelength:.3e}
Result:   f = {freq:.3e} Hz"""
        elif "planck" in q and "energy" in q:
            freq = extract_number(question)
            if freq is None:
                return "Please provide frequency in Hz."
            energy_j = PLANCK_H * freq
            return f"""📐 Planck Energy (E = hf)
Formula:  E = h × f
          E = 6.626 × 10⁻³⁴ × {freq:.3e}
Result:   E = {energy_j:.3e} J"""
        elif "angular momentum" in q or "bohr orbit" in q:
            n = extract_number(question)
            if n is None:
                return "Please provide the quantum number n."
            n        = int(n)
            L        = n * PLANCK_HBAR
            return f"""📐 Bohr Angular Momentum (L = nℏ)
Formula:  L = n × ℏ
Where:    n = {n}
          ℏ = 1.055 × 10⁻³⁴ J·s
Step 1: L = {n} × 1.055 × 10⁻³⁴
Result: L = {L:.3e} J·s"""
        else:
            return """I can calculate the following quantum physics formulas:

1.  Photon energy          E = hf
2.  Photon energy          E = hc/λ
3.  de Broglie wavelength  λ = h/mv
4.  Hydrogen energy level  Eₙ = -13.6/n² eV
5.  Energy transition      ΔE = E_upper - E_lower
6.  Photoelectric effect   KE = hf - Φ
7.  Uncertainty principle  Δx·Δp ≥ ℏ/2
8.  Frequency              f = c/λ
9.  Planck energy          E = hf
10. Angular momentum       L = nℏ

Please ask a numerical question with values provided."""
    except Exception as e:
        return f"Calculator error: {str(e)} — please check your values and try again."
def extract_number(text: str):
    """Extract first scientific or decimal number from text."""
    text = (text.replace("×10⁻³⁴", "e-34")
                .replace("×10⁻³¹", "e-31")
                .replace("×10⁻¹⁰", "e-10")
                .replace("×10⁻¹⁵", "e-15")
                .replace("×10⁶",   "e6")
                .replace("×10⁸",   "e8")
                .replace("×10¹⁴",  "e14")
                .replace("×10¹⁵",  "e15"))
    pattern = r'[-+]?\d+\.?\d*[eE][-+]?\d+|[-+]?\d+\.?\d+'
    matches = re.findall(pattern, text)
    return float(matches[0]) if matches else None
def extract_all_numbers(text: str):
    """Extract all numbers from text."""
    text = (text.replace("×10⁻³⁴", "e-34")
                .replace("×10⁻³¹", "e-31")
                .replace("×10⁻¹⁰", "e-10")
                .replace("×10⁻¹⁵", "e-15")
                .replace("×10⁶",   "e6")
                .replace("×10⁸",   "e8")
                .replace("×10¹⁴",  "e14")
                .replace("×10¹⁵",  "e15"))
    pattern = r'[-+]?\d+\.?\d*[eE][-+]?\d+|[-+]?\d+\.?\d+'
    matches = re.findall(pattern, text)
    return [float(m) for m in matches]
def tool_node(state: QuantumState) -> dict:
    question    = state["question"]
    tool_result = quantum_calculator(question)
    print(f"  [tool] calculated result for: {question[:50]}")
    return {"tool_result": tool_result}
print("-----")
print("QUANTUM CALCULATOR TESTS")
print("-----")

tests = [
    "What is the energy of a photon at frequency 6e14 Hz?",
    "Calculate de Broglie wavelength of electron at 2e6 m/s",
    "What is the hydrogen energy level at n=3?",
    "What is the energy transition from n=3 to n=2?",
    "Calculate uncertainty in momentum if delta x is 1e-10 m",
    "What is the frequency of light with wavelength 500e-9 m?",
    "Calculate angular momentum for n=2",
]
for test in tests:
    print(f"\nQ: {test}")
    result = tool_node({"question": test})
    print(result["tool_result"])
    print("-" * 55)

print("\n✅ tool_node tests complete")
```

NODE 7 -> EVAL NODE

```python
# ── Node 7: Eval ───────────────────────────────────────────
FAITHFULNESS_THRESHOLD = 0.7
MAX_EVAL_RETRIES = 2
def eval_node(state: QuantumState) -> dict:
    answer  = state.get("answer", "")
    context = state.get("retrieved", "")[:500]
    # CHANGED: read retries properly
    retries = state.get("eval_retries", 0)
    if not context:
        print(f"  [eval] skipped — no retrieval context")
        return {"faithfulness": 1.0, "eval_retries": 0}
    if retries >= MAX_EVAL_RETRIES:
        print(f"  [eval] max retries {retries} reached → forcing save")
        return {"faithfulness": 1.0, "eval_retries": retries}

    prompt = f"""Rate the faithfulness of this answer.
Does the answer use ONLY information from the context?
Reply with ONLY a number between 0.0 and 1.0.
1.0 = fully faithful. 0.5 = some hallucination. 0.0 = not faithful.

Context: {context}
Answer: {answer[:300]}"""
    result = llm.invoke(prompt).content.strip()
    try:
        score = float(result.split()[0].replace(",", "."))
        score = max(0.0, min(1.0, score))
    except:
        score = 0.5
    new_retries = retries + 1
    gate = "✅ PASS" if score >= FAITHFULNESS_THRESHOLD \
           else "⚠️  RETRY"
    print(f"  [eval] faithfulness={score:.2f} {gate} "
          f"(retry {new_retries}/{MAX_EVAL_RETRIES})")
    return {"faithfulness": score, "eval_retries": new_retries}

def eval_decision(state: QuantumState) -> str:
    score   = state.get("faithfulness", 1.0)
    retries = state.get("eval_retries", 0)
    if retries >= MAX_EVAL_RETRIES:
        print(f"  [eval_decision] max retries {retries} → saving anyway")
        return "save"

    if score >= FAITHFULNESS_THRESHOLD:
        print(f"  [eval_decision] PASS → saving answer")
        return "save"

    print(f"  [eval_decision] RETRY → regenerating answer")
    return "answer"

print("=" * 30)
print("EVAL NODE TESTS")
print("=" * 30)

test_1 = {
    "answer": """Quantum superposition means a quantum system
    exists in multiple states simultaneously until measured.""",
    "retrieved": """[Quantum Superposition]
    Quantum superposition states that a quantum system can
    exist in multiple possible states simultaneously until
    a measurement is made.""",
    "eval_retries": 0
}
result_1 = eval_node(test_1)
print(f"\nTest 1 — Faithful answer:")
print(f"  faithfulness: {result_1['faithfulness']:.2f}")
print(f"  eval_retries: {result_1['eval_retries']}")

test_2 = {
    "answer":      "The energy is 3.97 × 10⁻¹⁹ J",
    "retrieved":   "",
    "eval_retries": 0
}
result_2 = eval_node(test_2)
print(f"\nTest 2 — No context (tool result):")
print(f"  faithfulness: {result_2['faithfulness']:.2f}")
# CHANGED: eval_retries stays 0 when skipped
print(f"  eval_retries: {result_2['eval_retries']} (expected: 0)")

# CHANGED: eval_retries starts at MAX so hard stop triggers
test_3 = {
    "answer":    "Superposition is like a coin spinning.",
    "retrieved": """[Quantum Superposition]
    Quantum superposition states that a quantum system
    exists in multiple states simultaneously.""",
    "eval_retries": 2
}
result_3 = eval_node(test_3)
print(f"\nTest 3 — Max retries hit:")
print(f"  faithfulness: {result_3['faithfulness']:.2f} (expected: 1.0 forced)")
print(f"  eval_retries: {result_3['eval_retries']} (expected: 2)")

print("\n✅ eval_node tests complete")
```

```python
def route_decision(state: QuantumState) -> str:
    """
    After router_node — decides which path to take.
    Reads state["route"] and returns node name.
    """
    route = state.get("route", "retrieve")
    if route == "tool":         return "tool"
    if route == "memory_only":  return "skip"
    return "retrieve"
def eval_decision(state: QuantumState) -> str:
    """
    After eval_node — retry answer or save and finish.
    Reads faithfulness score and retry counter.
    """
    score   = state.get("faithfulness", 1.0)
    retries = state.get("eval_retries", 0)

    if score >= FAITHFULNESS_THRESHOLD:
        print(f"  [eval_decision] PASS → saving answer")
        return "save"

    if retries >= MAX_EVAL_RETRIES:
        print(f"  [eval_decision] max retries reached → saving anyway")
        return "save"

    print(f"  [eval_decision] RETRY → regenerating answer")
    return "answer"

graph = StateGraph(QuantumState)
graph.add_node("memory",memory_node)
graph.add_node("router",router_node)
graph.add_node("retrieve",retrieval_node)
graph.add_node("skip",skip_retrieval_node)
graph.add_node("tool",tool_node)
graph.add_node("viz",viz_selector_node)
graph.add_node("answer",answer_node)
graph.add_node("eval",eval_node)
graph.add_node("save",save_node)
graph.set_entry_point("memory")
graph.add_edge("memory","router")
graph.add_edge("retrieve","viz")
graph.add_edge("skip","viz")
graph.add_edge("tool","viz")
graph.add_edge("viz","answer")
graph.add_edge("answer","eval")
graph.add_edge("save",END)
graph.add_conditional_edges(
    "router",
    route_decision,
    {
        "retrieve":"retrieve",
        "skip":"skip",
        "tool":"tool"
    }
)
graph.add_conditional_edges(
    "eval",
    eval_decision,
    {
        "answer":"answer",
        "save":"save"
    }
)
checkpointer = MemorySaver()
app = graph.compile(checkpointer=checkpointer)

print("✅ Graph compiled successfully!")
print()
print("Graph flow:")
print("  memory → router")
print("         → retrieve → viz → answer → eval → save → END")
print("         → skip     ↗")
print("         → tool     ↗")
print()
print("Nodes:", [
    "memory", "router", "retrieve",
    "skip", "tool", "viz",
    "answer", "eval", "save"
])
```

```python
# ── Ask Helper ─────────────────────────────────────────────
def ask(question: str, thread_id: str = "test") -> dict:
    config = {"configurable": {"thread_id": thread_id}}
    result = app.invoke(
        {
            "question":       question,
            "eval_retries":   0,    
            "tool_result":    "",   
            "retrieved":      "",   
            "viz_type":       "",   
            "sources":        [],   
            "faithfulness":   0.0,  
            "answer":         "",   
        },
        config=config
    )
    return result

print("Running smoke test...")
print()
result = ask("What is quantum superposition?", thread_id="smoke_test")
print()
print(f"Question: What is quantum superposition?")
print(f"Route: {result.get('route', '?')}")
print(f"Sources: {result.get('sources', [])[:2]}")
print(f"Viz: {result.get('viz_type', '')}")
print(f"Faithfulness: {result.get('faithfulness', 0):.2f}")
print(f"Retries: {result.get('eval_retries', 0)}")
print(f"Answer: {result.get('answer', '')[:200]}...")
print()
print("✅ Smoke test complete")
```
